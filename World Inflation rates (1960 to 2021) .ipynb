```python
df=df.replace(to_replace=np.nan ,value=0)
df
```

```python
df2 = df.drop('Country',axis=1)
df['Mean'] = df2.mean(axis=1)
df['Max'] = df2.max(axis=1)
df['Max_Infiation_Year']=df2.idxmax(axis=1)

df
```

```python
x= df['Country']
y= df['2021']
plt.figure(figsize=[60,10])
plt.plot(x,y,'r.-')
plt.xlabel('Countries', fontdict={'fontsize':20})
plt.ylabel('Inflation Rate %', fontdict={'fontsize':20})
plt.title('Country Inflation 2021', fontdict={'fontsize':30,'fontweight':'bold'})
plt.xticks(df.Country,rotation= 90)
plt.style.use('ggplot')
plt.show()
```
